```python

def word_finder(filename):
    f=open(filename,"r")
    contents = f.read().split(" ")
    length = 0
    for i in contents:
        if len(i)>length:
            length = len(i)
            longest_word = i
    return longest_word
x=word_finder("classes.txt")
x
```
